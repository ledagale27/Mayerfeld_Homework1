After merging customers and transactions, some rows had missing transaction data. This occurs because some customers exist in the customer table but never made a purchase. Since our analysis focuses on actual purchases, I dropped rows with missing transaction IDs to ensure all subsequent calculations only consider real transactions.

The transactions and customers tables were merged using a left join on customer_id to combine purchase data with customer attributes like country, which allows to filter transactions from Spain and compute the average purchase value.

```python
# Dropping rows where transaction info is missing (these are not real purchases)
df1_merged = df1_merged.dropna(subset = ["transaction_id"])

# Checking result
print(df1_merged.isna().sum())
```

```python
spain_purchase_avg = df1_merged[df1_merged["country"] == "Spain"]["purchase_value"].mean()
print(spain_purchase_avg)
```

```python
contry_purchase_avg = (
    df1_merged
    .groupby("country")["purchase_value"]
    .mean()
    .reset_index()
)

print(contry_purchase_avg)
```

The average purchase value for customers from Spain is 75.63. 

```python
colors = [
    "orange" if country == "Spain" else "gray"
    for country in contry_purchase_avg["country"]
]

plt.figure()

plt.bar(contry_purchase_avg["country"], contry_purchase_avg["purchase_value"], color = colors)

plt.xlabel("Country")
plt.ylabel("Average Purchase Value")
plt.title("Average Purchase Value by Country")

plt.xticks(rotation=45)

plt.show()
```

```python
#Comparing the average purchase from Spain against the rest of the countries
spain_value = contry_purchase_avg[contry_purchase_avg["country"] == "Spain"]["purchase_value"].values[0]

rest_avg = contry_purchase_avg[contry_purchase_avg["country"] != "Spain"]["purchase_value"].mean()

comparison = pd.DataFrame({
    "group": ["Spain", "Rest"],
    "avg_purchase": [spain_value, rest_avg]
})

print(comparison)
```

Q7. What is the average number of pages viewed for customers who made a purchase (converted)?

The result should be the same; however, in this case, there is a slight difference, as some sessions marked as converted might not have ended in an actual purchase. Q8. What is the average time spent on the site for purchasing vs non-purchasing customers?

```python
# Identifying purchasing customers
purchasing_customers = transactions["customer_id"].unique()
```

```python
# Splitting sessions
# Sessions from customers who made a purchase
purchasing_sessions = sessions[
    sessions["customer_id"].isin(purchasing_customers)
]

# Sessions from customers who never purchased
non_purchasing_sessions = sessions[
    ~sessions["customer_id"].isin(purchasing_customers)
]
```

```python
# Calculating the average

avg_purchasing = purchasing_sessions["time_on_site"].mean()
avg_non_purchasing = non_purchasing_sessions["time_on_site"].mean()

print(f'The average time spent on the site for purchasing customer: {avg_purchasing}')
print(f'The average time spent on the site for non-purchasing customer: {avg_non_purchasing}')
```

Pattern #3 User Experience

```python
# Filtering for sessions that actually led to a purchase
df_ux = pd.merge(sessions, transactions, on = "customer_id", how = "left")
df_success = df_ux[df_ux["transaction_id"].notnull()].copy()

# Group Page Views into Bins for a cleaner bar chart
#  Do people buy after 5 pages or 20?"
bins = [0, 5, 10, 15, 20, 100]
labels = ['1-5', '6-10', '11-15', '16-20', '21+']
df_success['page_bins'] = pd.cut(df_success['pages_viewed'], bins=bins, labels=labels)

# 3. Aggregating the data
purchase_counts = df_success['page_bins'].value_counts().sort_index()

# 4. Creating the Plot
plt.figure(figsize=(10, 6))

# Applying the Magma palette (using your 0.2 to 0.7 logic)
colors = plt.cm.magma(np.linspace(0.2, 0.7, len(purchase_counts)))

# Creating the Bar Plot
bars = plt.bar(purchase_counts.index, purchase_counts.values, color=colors, edgecolor='black')

# 5. Styling & Zoom
plt.title('Purchases by Pages Viewed', fontsize=14, fontweight='bold')
plt.ylabel('Total Purchases (Count)')
plt.xlabel('Pages Viewed per Session')

# Zoom y-axis to highlight the differences between the peaks
# We start the y-axis at 50% of the minimum value to "lift" the bars
plt.ylim(purchase_counts.min() * 0.8, purchase_counts.max() * 1.1)

# Add data labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

### Extra analysis

RFM Analysis

Recency (R) = how recently they purchased

Frequency (F) = how often they purchase

Monetary (M) = how much thehy spend

```python
# Creating RFM table
rfm = transactions.groupby('customer_id').agg({
    'purchase_date': lambda x: (snapshot_date - x.max()).days,  # Recency
    'transaction_id': 'count',                                  # Frequency
    'purchase_value': 'sum'                                     # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
# Scoring each metric (1-4 or 1-5 scale)
# Recency: lower = better
rfm['R_score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])

# Frequency & Monetary: higher = better
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

# Combine scores
rfm['RFM_score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
```

```python
#Segment customers
def segment(row):
    if row['RFM_score'] in ['444', '443', '434']:
        return 'Champions'
    elif row['R_score'] >= 3 and row['F_score'] >= 3:
        return 'Loyal Customers'
    elif row['R_score'] == 4 and row['F_score'] == 1:
        return 'New Customers'
    elif row['R_score'] <= 2 and row['F_score'] >= 3:
        return 'At Risk'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment, axis=1)
```

```python
rfm['Segment'].value_counts().plot(kind='bar')

plt.title("Customer Segments (RFM)")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")

plt.show()
```

RFM analysis was used to segment customers based on recency, frequency, and monetary value. 
- "Others" is by far the largest group; however, these are occasional buyers who don't fit into a specific category. They provide volume, but they aren't currently driving consistent growth.
- "Loyal customers" are the steady customers who often buy.
- "At risk" are good customers who haven't purchased recently.
- "Champions" are the best customers, they have bought recently, often, and spend the most. They contribute in a high percentage of the total purchase value.
- "New customers" is the smallest segment; this suggests a low rate of recent customer acquisition compared to the existing base. The numbers suggest the business relies more on its historic customers than on fresh growth opportunities.

Key takeaways
- Addressing the "At risk" group as soon as possible may be a good move. A discount or re-engagement campaign could provide a boost.
- Focus on retention over acquisition behaviour shown."Loyal" and "Champion" segments are in a good place, but the massive "others" and "at risk" categories suggest that keeping customers engaged after the first purchase might be a challenge to overcome.
